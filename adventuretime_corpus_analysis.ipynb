# Adventure Time Linguistic Analysis
This notebook processes Adventure Time scripts to analyze word frequency and linguistic patterns across episodes.

```python

import os
from collections import Counter
import matplotlib.pyplot as plt
```

## 1. Setup and Imports

This section imports required Python libraries and prepares the environment for text processing. We use `Counter` for frequency analysis and `matplotlib` for visualization.

```python
scripts_folder = r"C:\Users\Tara\Documents\school-apps\GitHub\AdventureTime\script\seasonOne"

scripts = []

for filename in sorted(os.listdir(scripts_folder)):
    if filename.startswith("ATE") and filename.endswith(".txt"):
        with open(os.path.join(scripts_folder, filename), "r", encoding="utf-8") as f:
            scripts.append(f.read())

print(f"Loaded {len(scripts)} Season 1 scripts")
```

## 2. Loading Script Modules

This loop imports up to 26 Adventure Time script modules. Each module represents processed dialogue data for a specific episode. Successfully loaded modules are stored for later text processing.

```python
processed_texts = scripts
```

## 3. Text Processing

Each script is read as a raw text file and passed through preprocessing functions. These functions clean and normalize the text for analysis. All processed scripts are stored in a list for corpus creation.

```python

corpus = " ".join(processed_texts)

words = corpus.split()
word_counts = Counter(words)

top_words = word_counts.most_common(20)
words_plot, counts_plot = zip(*top_words)

plt.figure(figsize=(12,6))
plt.bar(words_plot, counts_plot)
plt.xticks(rotation=45)
plt.title("Top 20 Words in Adventure Time Season 1")
plt.show()
```

We combine all processed scripts into a single corpus. We then calculate word frequency using Python’s `Counter` and visualize the most common words using a bar chart.

# Season 10 Linguistic Analysis

```python
import os
from collections import Counter
import matplotlib.pyplot as plt

```

```python
scripts_folder = r"C:\Users\Tara\Documents\school-apps\GitHub\AdventureTime\script\seasonTen"

scripts = []

for filename in sorted(os.listdir(scripts_folder)):
    if filename.startswith("AT10E") and filename.endswith(".txt"):
        with open(os.path.join(scripts_folder, filename), "r", encoding="utf-8") as f:
            scripts.append(f.read())

print(f"Loaded {len(scripts)} Season 10 scripts")
```

```python
processed_texts = scripts
```

```python
corpus = " ".join(processed_texts)

words = corpus.split()
word_counts = Counter(words)

top_words = word_counts.most_common(20)

if top_words:
    words_plot, counts_plot = zip(*top_words)

    plt.figure(figsize=(12,6))
    plt.bar(words_plot, counts_plot)
    plt.xticks(rotation=45)
    plt.title("Top 20 Words in Adventure Time Season 10")
    plt.show()
else:
    print("No words found — check script loading")
```

## 5. Interpretation

Season 1’s most frequent words emphasize character identity and interpersonal dialogue, while Season 10’s top words shift toward functional language and narrative structure, reflecting the show’s evolution from character introduction to complex storytelling.
