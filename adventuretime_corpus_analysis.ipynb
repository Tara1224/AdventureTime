```python
import sys
import os
from collections import Counter
import importlib
import matplotlib.pyplot as plt

scripts_folder = r"C:\Users\Tara\Documents\school-apps\GitHub\AdventureTime\script"
sys.path.append(scripts_folder)

scripts = []
for i in range(1, 27):
    module_name = f"ATE{i}" if i > 1 else "ATE"
    try:
        mod = importlib.import_module(module_name)
        scripts.append(mod)
        print(f"Imported {module_name}")
    except ModuleNotFoundError:
        print(f"{module_name} not found")

processed_texts = []

for filename in os.listdir(scripts_folder):
    if filename.endswith(".txt"):
        with open(os.path.join(scripts_folder, filename), "r", encoding="utf-8") as f:
            text = f.read()
            for script in scripts:
                if hasattr(script, "process_text"):
                    text = script.process_text(text)
            processed_texts.append(text)

print(f"Processed {len(processed_texts)} scripts")

corpus = " ".join(processed_texts)
words = corpus.split()
word_counts = Counter(words)

top_words = word_counts.most_common(20)
print("Top 20 words:")
for w, c in top_words:
    print(f"{w}: {c}")

words_plot, counts_plot = zip(*top_words)
plt.figure(figsize=(12,6))
plt.bar(words_plot, counts_plot, color="#47A189")
plt.xticks(rotation=45)
plt.title("Top 20 Words in Adventure Time Season 1")
plt.show()
```
